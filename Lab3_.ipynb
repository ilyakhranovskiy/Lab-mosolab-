```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img_half = cv2.imread("Lab3_chess100.png", cv2.IMREAD_GRAYSCALE)
img_chess = cv2.imread("Lab3_half100.png", cv2.IMREAD_GRAYSCALE)

hist_half = cv2.calcHist([img_half], [0], None, [256], [0, 256])
hist_chess = cv2.calcHist([img_chess], [0], None, [256], [0, 256])

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_half, cmap = 'gray');
plt.subplot(122)
plt.plot(hist_half); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(img_chess, cmap = 'gray');
plt.subplot(122)
plt.plot(hist_chess); plt.xlim([0, 255]); plt.grid(); plt.show();

kernel3 = np.ones((3,3),np.float32) / (3 * 3)
kernel5 = np.ones((5,5),np.float32) / (5 * 5)
kernel9 = np.ones((9,9), np.float32) / (9 * 9)
kernel15 = np.ones((15,15), np.float32) / (15 * 15)

blur_half3= cv2.filter2D(img_half, -1, kernel3)
blur_half5= cv2.filter2D(img_half, -1, kernel5)
blur_half9= cv2.filter2D(img_half, -1, kernel9)
blur_half15= cv2.filter2D(img_half, -1, kernel15)

blur_chess3 = cv2.filter2D(img_chess, -1, kernel3)
blur_chess5 = cv2.filter2D(img_chess, -1, kernel5)
blur_chess9 = cv2.filter2D(img_chess, -1, kernel9)
blur_chess15 = cv2.filter2D(img_chess, -1, kernel15)

blur_half3_hist= cv2.calcHist([blur_half3], [0], None, [256], [0, 256])
blur_half5_hist= cv2.calcHist([blur_half5], [0], None, [256], [0, 256])
blur_half9_hist= cv2.calcHist([blur_half9], [0], None, [256], [0, 256])
blur_half15_hist= cv2.calcHist([blur_half15], [0], None, [256], [0, 256])

blur_chess3_hist=cv2.calcHist([blur_chess3], [0], None, [256], [0, 256])
blur_chess5_hist=cv2.calcHist([blur_chess5], [0], None, [256], [0, 256])
blur_chess9_hist= cv2.calcHist([blur_chess9], [0], None, [256], [0, 256])
blur_chess15_hist=cv2.calcHist([blur_chess15], [0], None, [256], [0, 256])


plt.figure(figsize=(20,20))
plt.subplot(121)
plt.plot(blur_half3_hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.plot(blur_half5_hist); plt.xlim([0, 255]); plt.grid(); plt.show();

plt.figure(figsize=(20,20))
plt.subplot(121)
plt.imshow(blur_chess3_hist, cmap = 'gray');
plt.subplot(122)
plt.plot(blur_chess5_hist); plt.xlim([0, 255]); plt.grid(); plt.show();



```
